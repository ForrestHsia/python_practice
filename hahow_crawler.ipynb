```python
import requests,time
```

```python
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36 Edg/93.0.961.38'
}
```

```python
course = []
for index in range(3):
    hahow_url = "https://api.hahow.in/api/courses?limit=24&page="
    hahow_url = hahow_url + str(index)
    hahow_result = requests.get(hahow_url, headers=headers)
    print(hahow_result)
    time.sleep(10)
    hahow_resultJson = hahow_result.json()
    
    for data in hahow_resultJson['data']:
        course.append(data['title'])
        course.append(data['owner']['name'])
        course.append(data['price'])
        course.append(data['preOrderedPrice'])
        course.append(data['numSoldTickets'])
```
